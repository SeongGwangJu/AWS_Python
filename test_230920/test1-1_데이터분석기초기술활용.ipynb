```python
#문제1
carNumberInput = input("차량 번호 입력 :")

# 맨 앞에 3자리 숫자를 입력했는지 검증
try:
    typeValue = int(carNumberInput.split(" ")[0][0:3])
except ValueError:
    typeValue =int(carNumberInput.split(" ")[0][0:2]) #2자리 숫자 입력시에도 '오류'결과 출력을 위해
type=""

# 차량용도 저장
purposeValue = carNumberInput.split(" ")[0][-1]
purpose=""

# 일련번호 숫자입력 검증
try:
    number = int(carNumberInput.split(" ")[1])
except ValueError:
    print("뒷자리에는 숫자만 입력해주세요 정상작동하지 않습니다.")

# 앞/뒤 자릿수 검증
if len(carNumberInput.split(" ")[0]) !=4: # 앞자리가 4자리가 아니면 모두 오류
    type = '오류'
    purpose ='오류'
elif len(carNumberInput.split(" ")[1]) !=4: #뒷자리가 4자리가 아니면
    number = '오류'
else: #형식이 올바른 경우
    #차량종류 판별 : 앞에서 3자리 숫자임을 검증함 : 100 ~ 999
    if typeValue < 700:
        type = '승용차'
    elif typeValue < 800:
        type = '승합차'
    elif typeValue < 980:
        type = '화물차'
    elif typeValue < 998:
        type = '특수차'
    else:
        type = '긴급차'

    #차량용도 판별 #그냥 or하면 안되길래..
    if (purposeValue == '가' or purposeValue == '나' or purposeValue == '다' or purposeValue == '라' or purposeValue == '마' or
        purposeValue == '거' or purposeValue == '너' or purposeValue == '더' or purposeValue == '러' or purposeValue == '머' or
        purposeValue == '버' or purposeValue == '서' or purposeValue == '어' or purposeValue == '저' or purposeValue == '고' or
        purposeValue == '노' or purposeValue == '도' or purposeValue == '로' or purposeValue == '모' or purposeValue == '보' or
        purposeValue == '소' or purposeValue == '오' or purposeValue == '조' or purposeValue == '구' or purposeValue == '누' or
        purposeValue == '두' or purposeValue == '루' or purposeValue == '무' or purposeValue == '부' or purposeValue == '수' or
        purposeValue == '우' or purposeValue == '주'):
        purpose = '비상업용'
    elif purposeValue == '아' or '바' or '사' or '자':
        purpose = '운수사업용'
    elif purposeValue == '배':
        purpose = '택배차'
    elif purposeValue == '허' or '하' or '호':
        purpose = '렌터카'
    else:
        purpose = '오류'

outputDic = {
        "type":typeValue,
        "purpose":purposeValue,
        "number":number
    }

carNumberInfo = {
        "type":type,
        "purpose":purpose,
        "number":number
    }

print(carNumberInput)
print(outputDic)
print(carNumberInfo)


```
