```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://pages.coupang.com/p/96636"

response = requests.get(url)

html = bs(response.text, 'html.parser')
#방법1 : select : js에서 쿼리셀렉터같은 애
imgList = []
imgs = html.select(".sale_point_product__product_image > img")
for img in imgs:
    if bool(img.get("lazy-load")):
        imgLink = "https://" + img.get("lazy-load")[2:]
        imgList.append(imgLink)
print(imgList)

#방법2 : find_all : js에서 겟엘리먼트같은 애
imgList = []
divs = html.find_all("div", class_="sale_point_product__product_image")
for div in divs:
    img = div.find("img")
    # if bool(img.get("lazy-load")):
    if img and img.get("lazy-load"):
        # print(img.get("lazy-load"))
        # print("https://" + img.get("lazy-load")[2:])
        imgLink = "https://" + img.get("lazy-load")[2:]
        imgList.append(imgLink)
print(imgList)
```

```python
#이미지, 제품, 가격 가져오기(1)
import requests
from bs4 import BeautifulSoup as bs

response = requests.get('https://pages.coupang.com/p/96636').text
productInfo = {}

html = bs(response, "html.parser")

productDomList = html.select(".sale_point_product.grid-2")
productDomList
productInfoList = []
_img= None
_title= None
_price = None
for product in productDomList:
  try:
    for img in product.select(".sale_point_product__product_image img"):
      _img = "http:"+img['lazy-load']
    for title in product.select(".info_section__title"):
      _title = title.text
    for price in product.select("prict_info span"):
      _price = price.text
    productInfo = {
      "img": _img,
      "title": _title,
      "price":_price
    }
    productInfoList.append(productInfo)
  except Exception as e:
    print(e)
productInfoList
```

```python
#이미지, 제품, 가격 가져오기(2)
import requests
from bs4 import BeautifulSoup as bs

response = requests.get('https://pages.coupang.com/p/96636').text
productInfo = {}

html = bs(response, "html.parser")

#상품들 안에 각각의 상품 div들
sale_point_products = html.select(".sale_point_products .sale_point_product")


productList = []

for product in sale_point_products:
    productName = product.select_one(".info_section__title").text
    price = product.select_one(".price_info__discount")
    price = price.text if price != None else "" # 3항연산자
    src = product.select_one(".sale_point_product__product_image > img").get("lazy-load")
    src = "https:" + src if src != None else ""

    productDict = {
        "productName" : productName,
        "price" : price,
        "src": src
    }

    productList.append(productDict)

productList
```
